### - Get all Pokemons whose spawn time is less than 300 seconds.

```python
df2 = df1.copy()
```

```python
def time_seconds(val): 
    if type(val)==str:
        #print(val)
        #print(type(val))
        s1 = int(val.split(':')[0])
        s2 = int(val.split(':')[1])
        #print(s1,s2)
        secs = s1*60 + s2
        #print(secs)
        return secs
    return 0

df2['Spawn_time_in_seconds'] = df2['spawn_time'].apply(time_seconds)
```

```python
df2[df2['Spawn_time_in_seconds']<300]
```
